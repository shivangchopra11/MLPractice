```python
def PCA(data, m=None):
    if m is None:
        m = data.shape[1]
    cov = np.cov(data, rowvar=False)
    eig_val, eig_vec = np.linalg.eig(cov)
    new_pts = data.dot(eig_vec)
    red_data = data[:,-m-1:]
    return eig_val, eig_vec,data,red_data
```
