```python
titles= ds.get('movie_title')
print titles
print titles.shape
print(type(titles))
```

```python
freq_titles = dict()
for title in titles:
    leng = len(title)
    if leng in freq_titles:
        freq_titles[leng] += 1
    else:
        freq_titles[leng]=1
```

```python
plt.plot(freq_titles.keys(),freq_titles.values())
plt.show()
```

```python
len_titles = [len(i) for i in titles]
print len(len_titles)
```
